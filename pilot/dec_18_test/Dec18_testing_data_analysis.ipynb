```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
import warnings 
warnings.filterwarnings('ignore')
```

---

```python
during=pd.read_csv("during_cpt.csv")
during
```

```python
during.describe()
```

```python
plt.figure(figsize=(10,8))
plt.title("HRV during CPT")
plt.plot(during['Time'],during['HRV'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("GSR during CPT")
plt.plot(during['Time'],during['GSR'])
plt.show()
```

---

```python
pre=pd.read_csv("pre.csv")
pre
```

```python
pre.describe()
```

```python
plt.figure(figsize=(10,8))
plt.title("HRV before CPT")
plt.plot(pre['Time'],pre['HRV'],color="red")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("GSR before CPT")
plt.plot(pre['Time'],pre['GSR'], color="red")
plt.show()
```

---

```python
post=pd.read_csv("post.csv")
post
```

```python
post.describe()
```

```python
plt.figure(figsize=(10,8))
plt.title("HRV after CPT")
plt.plot(post['Time'],post['HRV'],color="green")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("GSR after CPT")
plt.plot(post['Time'],post['GSR'],color="green")
plt.show()
```

---

```python
x = np.linspace(1,74,num=74)
y1 = during['HRV'].head(74)
y2 = post['HRV'].head(74)
```

```python
plt.figure(figsize=(10,8))
plt.title('HRV Changes')
plt.plot(x,pre['HRV'],label='pre-CPT')
plt.plot(x,y1,label='during-CPT')
plt.plot(x,y2,label='post-CPT')
plt.legend()
plt.show()
```

```python
y3 = during['GSR'].head(74)
y4 = post['GSR'].head(74)
```

```python
plt.figure(figsize=(10,8))
plt.title('GSR Changes')
plt.plot(x,pre['GSR'],label='pre-CPT')
plt.plot(x,y3,label='during-CPT')
plt.plot(x,y4,label='post-CPT')
plt.legend()
plt.show()
```

```python
df_all=pd.concat([pre,during,post],keys=['pre', 'during','post'])
df_all
```

```python
plt.figure(figsize=(10,8))
plt.title('GSR Changes')
x=np.linspace(0,249,250)
y=df_all['GSR']
plt.plot(x,y)
plt.axvline(x = 73, color = 'grey', label = 'axvline - full height',linestyle=':')
plt.axvline(x = 163, color = 'grey', label = 'axvline - full height',linestyle=':')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('HRV Changes')
x=np.linspace(0,249,250)
y=df_all['HRV']
plt.plot(x,y)
plt.axvline(x = 73, color = 'grey', label = 'axvline - full height',linestyle=':')
plt.axvline(x = 163, color = 'grey', label = 'axvline - full height',linestyle=':')
plt.show()
```
